```python
#1st Problem Statement
a = int(input("Marks pf subject one = "))
b = int(input("Marks of subject two = "))
c = int(input("Marks of subject three = "))
d = int(input("Marks of subject four = "))
e = int(input("Marks of subject five = "))

A = a+b+c+d+e
B = A/5

print(A)
print(B)
```

```python
#2nd Problem Statement
r = int(input("Enter the radius of circle: "))
Area = 3.14*r*r
print(Area)
```

```python
#3rd Problem Statement
l = int(input("Enter the length of rectangle: "))
b = int(input("Enter the breadth of rectangle: "))
Area = l*b
print(Area)
```

```python
#4th Problem Statement
l = int(input("Enter the length of rectangle: "))
b = int(input("Enter the breadth of rectangle: "))
Perimeter = 2*(l+b)
print(Perimeter)
```

```python
#Simple Calculator (5th Problem Statement)
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
sum=a+b
multiplication = a*b
subtraction = a-b
division = a/b
print(sum,subtraction,division,multiplication)
```
